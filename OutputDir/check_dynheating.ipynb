```python
#Import modules

import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
#%matplotlib inline
#%matplotlib widget
import cartopy.crs as ccrs
from ipywidgets import interact, IntSlider, SelectionSlider, Dropdown
```

```python
#Open file and list variables
filename = "/home/ucfacb0/rundirs/Rockets/gc_4x5_47L_merra2_fullchem_main/OutputDir/Baseline/GEOSChem.DynHeating.20200101_0300z.nc4"
ds = xr.open_dataset(filename, decode_times=False)
ds
```

```python
varnames = list(ds.data_vars.keys())
max_alt = ds["DynHeating"].shape[1]
xstep = 5
ystep = 4
xmin  = ds.coords["lon"].values[0] - xstep/2
xmax  = ds.coords["lon"].values[-1] + xstep/2
ymin  = ds.coords["lat"].values[0] - ystep/2
ymax  = ds.coords["lat"].values[-1] + ystep/2
print(xmin,xmax,xstep)

#Define coordinates

def plot_layer(l):
    fig = plt.figure(figsize=[8, 4])
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.coastlines()
    
    ax.set_xticks(np.linspace(ds.coords["lon"].values[0],-ds.coords["lon"].values[0],5),minor=False)
    ax.set_yticks(np.linspace(ds.coords["lat"].values[0],-ds.coords["lat"].values[0],5),minor=False)
    #ax.gridlines(xlocs=np.linspace(xmin, xmax, len(ds.coords["lon"].values)+1), ylocs=np.linspace(ymin, ymax, len(ds.coords["lat"].values)+1))
    ds["DynHeating"].isel(lev=l).plot(ax=ax)
    ax.set_title('DynHeating; alt={l}')

interact(plot_layer,  
         l=IntSlider(min=0, max=max_alt-1, step=1, continuous_update=False)
        );
```
