```python
# Import final spinup file
spinup_file = f"/home/ucfacb0/rundirs/Rockets/gc_4x5_47L_merra2_fullchem_main/Restarts/Spinup/GEOSChem.Restart.20200101_0000z.nc4"
ds_spinup = xr.open_dataset(spinup_file)
new_species = ['SpeciesRst_PSO4AQ', 'SpeciesRst_PH2SO4', 'SpeciesRst_TSOG3', 'SpeciesRst_TSOG2',
               'SpeciesRst_TSOG1', 'SpeciesRst_TSOG0', 'SpeciesRst_TSOA3', 'SpeciesRst_TSOA2',
               'SpeciesRst_TSOA1', 'SpeciesRst_TSOA0', 'SpeciesRst_FURA', 'SpeciesRst_BUTDI',
               'SpeciesRst_ASOG3', 'SpeciesRst_ASOG2', 'SpeciesRst_ASOG1', 'SpeciesRst_ASOAN',
               'SpeciesRst_ASOA3', 'SpeciesRst_ASOA2', 'SpeciesRst_ASOA1']

varnames = list(ds_spinup.data_vars.keys())
varname_list, sample_list, spinup_list, diff_list, diffpercent_list,new_list = [],[],[],[],[],[]
for varname in varnames:
    varname_list.append(varname)
    sample_list.append(np.sum(ds_v14_3_0[varname].values))
    spinup_list.append(np.sum(ds_spinup[varname].values))
    diff_list.append(np.abs(np.sum(ds_spinup[varname].values)-np.sum(ds_v14_3_0[varname].values)))
    diffpercent_list.append((np.abs(np.sum(ds_spinup[varname].values)-np.sum(ds_v14_3_0[varname].values)))/(np.sum(ds_v14_3_0[varname].values)*100))
    if varname in new_species:
        new_list.append("1")
    else:
        new_list.append("0")
        

data = pd.DataFrame({'varname': varname_list, 'sample_list': sample_list, 'spinup_list': spinup_list, 
                     'diff_list': diff_list, 'diffpercent_list': diffpercent_list, 'new_list': new_list},
                    columns=['varname', 'sample_list', 'spinup_list', 'diff_list', 'diffpercent_list', 'new_list'])

def make_pretty(styler):
    styler.background_gradient(axis=1, vmin=0, vmax=1, cmap="YlGnBu",subset="diffpercent_list")
    styler.format('{:.4e}', subset=['sample_list', 'spinup_list', 'diff_list'])
    styler.format('{:.4e}', subset=['diffpercent_list'])
    return styler


#data.to_csv(f"{file}.csv",sep=',')
data = data.sort_values(["new_list"], ascending=False)
#data[data["diff_list"] != None].style.pipe(make_pretty)
#ds_spinup["SpeciesRst_AL2O3"]
```
